```python
import sys

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model

regr = linear_model.LinearRegression()
sns.set(style="white")
sns.set_context("paper", 1.5)
```

```python
df = pd.read_csv("mw-ob-new.csv")

X = df['MW'].values.reshape(-1, 1)
y = df['RMSD'].values.reshape(-1, 1)
regr.fit(X, y)
print(regr.coef_[0])
print(regr.intercept_)


g = sns.jointplot(df['MW'], df['RMSD'], kind="kde")
sns.regplot(df['MW'], df['RMSD'], ax=g.ax_joint, scatter=False)
g.ax_joint.set_xlabel("Molecular weight (Da)")
g.ax_joint.set_ylabel("RMSD ($\AA$)")
plt.xlim([100, 650])
plt.ylim([-0.5, 4.5])
plt.tight_layout()
plt.savefig("mw-ob-new.pdf")
```
