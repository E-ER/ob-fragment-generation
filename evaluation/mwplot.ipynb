```python
import sys

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set(style="white")
sns.set_context("paper", 1.8)
```

```python
df = pd.read_csv("mw-ob-new.csv")

slope, intercept, r_value, p_value, std_err = stats.linregress(df['MW'],df['RMSD'])

g = sns.jointplot(df['MW'], df['RMSD'], kind="reg", joint_kws={'color':'k'}, 
                  scatter=False, xlim=(100, 650), ylim=(0, 5))
sns.kdeplot(df['MW'], df['RMSD'], ax=g.ax_joint, cmap='Blues')
g.ax_joint.set_xlabel("Molecular weight (Da)")
g.ax_joint.set_ylabel("RMSD ($\AA$)")
#g.ax_joint.legend_.remove()
g.ax_joint.legend_.texts[0].set_text("$R^2$: {:.2f}, y={:.3f}x{:.3f}".format(r_value**2, slope, intercept))
plt.tight_layout()
plt.savefig("mw-ob-new.pdf")
```
